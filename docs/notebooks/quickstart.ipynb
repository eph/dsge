# DSGE Quickstart Notebook

Minimal example: load a model, compile it, and compute IRFs.

```python
from dsge import read_yaml

# Load a tiny AR(1) example bundled with the repo
m = read_yaml('dsge/examples/ar1/ar1.yaml')
print(m)
```

```python
# Compile and get system matrices
lin = m.compile_model()
p0 = m.p0()
CC, TT, RR, QQ, DD, ZZ, HH = lin.system_matrices(p0)
TT.shape
```

```python
# Simple IRFs for 20 periods
irfs = lin.impulse_response(p0, h=20)
list(irfs.keys())
shock = list(irfs.keys())[0]
irfs[shock].head()
```
