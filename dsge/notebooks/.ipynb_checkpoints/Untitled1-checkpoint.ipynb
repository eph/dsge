```
from dsge.translate import smc
from fortress import make_smc
import numpy as np

from dsge.translate import write_prior_file
```

```
from dsge.DSGE import DSGE
nkmp = DSGE.read('/home/eherbst/Dropbox/from-work/small-example/dsge1.yaml')
from dsge.DSGE import DSGE
mod = nkmp.compile_model()

np.savetxt('_fortress_tmp/yy.txt',mod.yy)
write_prior_file(mod.prior, '_fortress_tmp/')
model_file = smc(nkmp)
model = make_smc(model_file, other_files={})
```

```
res = model.run(npart=1000, nphi=200, bend=2.0, nproc=4)
```
